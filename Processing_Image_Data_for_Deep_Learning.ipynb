**Convert RGB Images to Grayscale image using OpenCV**

```python
# importing OpenCV library

import cv2
```

```python
img=cv2.imread("/content/OIP.lYAJIsRV41BwpukJEZBU8QHaEK_rs=1&pid=ImgDetMain&o=7&rm=3")
```

```python
type(img)
```

```python
img.shape
```

```python
# must be done on numpy array format

grayscale_image =cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
```

```python
type(grayscale_image)
```

```python
grayscale_image.shape  # now it will not be 3 channel, it will just show the dimensions
```
